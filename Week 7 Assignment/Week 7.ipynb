```python
import numpy as np, cv2
from matplotlib import pyplot as plt

#A)

matrix_a = np.array([
    [1, -2, 3, 7],
    [2, 1, 1, 4],
    [-3, 2, -2, 10]
])

matrix_b = np.array([
    [0, 1, -3, -2],
    [10, -1, 2, -3],
    [5, 1, -1, 4]
])

matrix_c = np.array([
    [4, 0, -2, 3],
    [3, 6, 9 ,7],
    [2, 2, 5, 1],
    [9, 4, 6, -2]
])

print(f'2) \n Matrix A: {matrix_a.shape} \n Matrix B: {matrix_b.shape} \n Matrix C: {matrix_c.shape}')
```

```python
a_sum_b = matrix_a + matrix_b
a_mult_b = matrix_a * matrix_b
c_transpose = matrix_c.transpose()
print(f'3) \n A+B: {a_sum_b} \n AB: {a_mult_b} \n AC: AC cannot be defined due to the two having differing shapes. \n C Transposed: {c_transpose}')
```

```python
print(f'''4: {a_sum_b.transpose() == matrix_a.transpose() + matrix_b.transpose()} \n 5: You cannot perform this transposistion using those two matrices due to unmatched shapes. \n
6: {np.power(matrix_c.astype(float), -1)}''')
```

```python
import numpy as np
from matplotlib import pyplot as plt

#Setup the matrix table
coef_values = np.array([[27, -10, 4, -29],
                        [-16, 5, -2, 18],
                        [-17, 4, -2, 20],
                        [-7, 2, -1, 8]])
result_values = np.array([1, -1, 0, 1])

#Solve values
solved_values = np.linalg.solve(coef_values, result_values)

#Display solved table
display(solved_values)

#Test the solved table for each equation and print a truth table
#1)
print([(27*10 - 10*3 + 4*(-9) - 29*7) == 1,
(-16*10 + 5*3 - 2*(-9) + 18*7) == -1,
(-17*10 + 4*3 - 2*-9 + 20*7) == 0,
(-7*10 + 2*3 - (-9) + 8*7) == 1])
```

```python
vector_transformation_matrix = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
                                [np.sqrt(2)/2, np.sqrt(2)/2]])
transform_func = lambda x: vector_transformation_matrix*x

defined_matrix = np.array([[2], [-2]])

#2A)
print(transform_func(defined_matrix))
```

```python
#2A)
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)

xy = np.vstack([xx.flatten(), yy.flatten()])
transform_matrix_plt = vector_transformation_matrix @ xy

xx_transformed = transform_matrix_plt[0].reshape(xx.shape)
yy_transformed = transform_matrix_plt[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
```

```python
import pandas as pd, numpy as np, sympy, matplotlib.pyplot as plt, matplotlib.patches as patches

#Set plot style for future use
plt.rcParams.update(plt.rcParamsDefault)
plt.style.use('seaborn-v0_8-dark')

def reimann_sum(formula:str, a:int, b:int, n:int, debug:bool=False, sum_val:bool = True):
    '''calculate the riemann sum on a given interval using the right-hand rule. \n
    F is the formula to use, where the variable is X. 
    A and B are defined for given interval [a,b] \n
    N is the number of rectangles to calculate
    If an error occurs None is returned instead. Error messages are suppressed by default, pass bool=True to enable.
    Returns the reinmann sum by default, you can return the list of Rn values by setting sum_val=False
    '''
    #Attempt to process normally, catching any exception and optionally printing them. 
    try:
        #Catch any negative or 0 value that would cause processing or accurcy issues and return early
        if n <= 0:
            if debug:
                print("The number of rectangles must be greater than 0")
            return None
        #Take the formula from the input string, literally evaluate it, and use lambda x to allow for f(x) processing later
        f = lambda x: eval(formula)
        #Find rectangle width
        rect_width = (b-a)/n
        #create a list comprehension of f(a+k*x) values then sum them together if needed. Return after 
        f_values = [f(a + i*rect_width) for i in range(1,n+1)]
        if sum_val:
            curve_area = sum(f_values)
            return curve_area
        else:
            return f_values
    except Exception as e:
        if debug:
            print(e)
        return None
print(f'Testing function for formula 2x with interval [1,4] and 4 rectangles: {reimann_sum("2*x", 1,4,4)}')

```

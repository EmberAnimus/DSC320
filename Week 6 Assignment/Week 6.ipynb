```python
import pandas as pd, numpy as np, pathlib, os
from pathlib import Path

cur_path = Path()
os.chdir(cur_path)

#1)
def norm_data(data:list):
    'Takes in a list of data and normalizes it. In the event that the data max and min are the same it returns a ZeroDivisionError'
    data_min = min(data)
    data_max = max(data)
    if data_max-data_min == 0:
        print("Your dataset minimum and maximum are the same value, data can't be normalized as its all the same.")
        return ZeroDivisionError
    data = [(i-data_min)/(data_max-data_min) for i in data]
    return data

#2)
def stand_data(data:list):
    'Takes a list of data and standardizes it.'
    data_mean = np.mean(data)
    data_std_dev = np.std(data)
    data = [(i-data_mean)/data_std_dev for i in data]
    return data

ca_housing_data = pd.read_csv('calif_housing_data.csv')
display(ca_housing_data)
```

3C) As seen below, the feature we create is indicative of the average number of bedrooms per given household.

```python
#3C)
ca_housing_data['avg_bedrooms'] = ca_housing_data['total_bedrooms']/ca_housing_data['households']
display(ca_housing_data)
```

```python
#3D)
ca_housing_data_slice = ca_housing_data[['housing_median_age', 'median_income', 'avg_bedrooms']]
display(ca_housing_data_slice)

#3E)
ca_housing_data_slice = ca_housing_data_slice.apply(lambda x: stand_data(x))
display(ca_housing_data_slice)
```

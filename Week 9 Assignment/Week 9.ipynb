```python
import numpy as np, pandas as pd, os, matplotlib as mpl
from pathlib import Path
from matplotlib import pyplot as plt

os.chdir(Path())

vg_data = pd.read_csv('video_game_data.csv')
```

```python
def check_eig(a, eig_val, eig_mtx):
    result_mtx = [np.allclose(np.dot(a, eig_mtx[:,i]),np.dot(eig_val[i], eig_mtx[:,i])) for i in range(len(eig_mtx))]
    return result_mtx
mtx_a = np.array([[4, 0, 1],
                [-1, -6, -2],
                [5, 0, 0]])
ver_mtx_1 = np.array([1,2,3])
ver_mtx_2 = np.array([0,1,0])
eg_val, eg_mtx = np.linalg.eig(mtx_a)
print(f'Function test, should return all true: \t {check_eig(mtx_a, eg_val, eg_mtx)} \n')
print(f'Vector A: {ver_mtx_1.tolist()} {"is" if np.all(check_eig(mtx_a, ver_mtx_1, eg_mtx)) else "is not"} an eigenvalue of the matrix')
print(f'Vector B: {ver_mtx_2.tolist()} {"is" if np.all(check_eig(mtx_a, ver_mtx_2, eg_mtx)) else "is not"} an eigenvalue of the matrix')
print(f'\n Eigenvalues {eg_val} \n Eigenmatrix: {eg_mtx}')
```
